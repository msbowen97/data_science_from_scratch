```python
import numpy as np
import pandas as pd
from preprocessing import Method
```

```python
data = pd.read_csv('./spaceship-titanic/train.csv')
Xdata = data[[col for col in data.columns if col != 'Transported']]
yData = data['Transported'].astype(float)
```

```python
Xdata[['Deck','RoomNo','ShipSide']] = Xdata['Cabin'].str.split('/', expand=True)
Xdata[['GroupNo', 'GroupSize']] = Xdata['PassengerId'].str.split('_', expand=True)
Xdata['Child'], Xdata['Infant'] = (Xdata['Age'] <= 18).astype(int), (Xdata['Age'] <= 5).astype(int)
Xdata['TotalSpent'] = np.sum(Xdata[['RoomService', 'FoodCourt', 'ShoppingMall','Spa','VRDeck']], axis=1)
```

```python
method = Method(dropCols = ['Name', 'Cabin', 'PassengerId', 'RoomNo', 'GroupNo', 'FoodCourt', 'ShoppingMall'],
                onehotCols = ['HomePlanet', 'Destination', 'ShipSide'], directCols=['VIP', 'CryoSleep', 'GroupSize'],
                ordinalCols = {'Deck': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'T']}, inlineDiagnostics=True,
                splitMethod='kfold', splitGroups=5, metrics = ('accuracy', 'precision', 'recall',),
                imputationMethod='lazy', normalizationMethod='zscore', bagging=True)
```

```python
method(X=Xdata, y=yData, lossMethod='gini', modelMethod='decision_tree', maxTreeDepth=3, minGroupSize=5,
       nCols=1, nRows=1, nTrees=1)
```
